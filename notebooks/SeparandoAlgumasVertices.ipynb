Como visto no jupyter notebook `APS 1`, a visualização do grafo do DataSet não foi uma das melhores. Isso porque haviam muitas arestas, e um número de vértices considerável. Foi pensando em melhorar a visualização que então decidimos separar cerca de 2000 arestas e as vértices associadas a elas. Para isso, criamos novos arquivos de texto que contém as arestas, as vértices e a propriedade das vértices (os nomes das páginas de Facebook), que serão usados para a construção de um grafo menor - o que pode ser visto no jupyter notebook `GrafoMenor`.

Para isso começando lendo os arquivos de arestas e vértices do DataSet original:

```python
# Lendo os arquivos e usando readlines()
fileArestas = open('./fb-pages-sport.edges', 'r')
fileVertices = open('./fb-pages-sport.nodes', 'r')
linhasArestas = fileArestas.readlines()
linhasVertices = fileVertices.readlines()
```

Agora, iremos criar uma lista de aproximadamente 2000 arestas que depois será transformada em um arquivo no formato `txt`:

```python
listaArestas = []
contador = 0
for linhaAresta in linhasArestas:
    linha1 = linhaAresta.strip().split(',')
    if contador == 1999:
        break
    else:
        par = [int(linha1[0]), int(linha1[1])]
        listaArestas.append(par)
    contador +=1
            
```

```python
#print(listaArestas) [debug]
```

Mais uma vez o processo é repetido, mas agora com as vértices e o nome delas.


**Observação**: pedimos para que **não rode esta célula abaixo**, isso porque o tempo para rodar é longo, uma vez que há iteração sobre todas as arestas selecionadas e sobre todas as vértices do DataSet (já que precisamos achar o label do nó que será adicionado na lista).

```python
listaVertices = []
listaNomeVertices = []

for aresta in listaArestas:
    for i in range (1,len(linhasVertices)):
        linha2 = linhasVertices[i].strip().split(',')     
        
        if int(aresta[0]) not in listaVertices and int(aresta[0]) == int(linha2[len(linha2)-1]):
            listaVertices.append(aresta[0])  
            listaNomeVertices.append(linha2[1])
            
        if int(aresta[1]) not in listaVertices and int(aresta[1]) == int(linha2[len(linha2)-1]):
            listaVertices.append(aresta[1])  
            listaNomeVertices.append(linha2[1])

            
```

```python
#print(listaNomeVertices) [debug] 

#print(len(listaNomeVertices)) [debug] -- para conferir se o número de propriedades das vértices equivale 
#print(len(listaVertices)) [debug] --     ao número de vértices selecionadas.
```

Por fim, os arquivos no formato `txt` foram criados com o conteúdo das três listas: `listaVertices`, `listaNomeVertices` e `listaArestas`:

```python
with open('verticesReduzidos.txt', 'w') as filehandle:
    for listitem in listaVertices:
        filehandle.write(str(listitem)+', ')

with open('nomeVerticesReduzidos.txt', 'w') as filehandle:
    for string in listaNomeVertices:
        filehandle.write(str(string)+'\n')
```

```python
with open('arestasReduzidas.txt', 'w') as filehandle:
    for listitem in listaArestas:
        filehandle.write(str(listitem)+'\n')
```
